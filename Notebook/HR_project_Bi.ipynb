```python
df = df.drop(columns=['EmployeeCount'])
```

```python
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include='object').columns
```

```python
df[num_cols] = df[num_cols].fillna(0)          # numeric safe
df[cat_cols] = df[cat_cols].fillna('Unknown')  # categorical safe
```

```python
df
```

```python
df.loc[df['Age'] == 0, 'Age'] = np.nan
df['Age'] = df['Age'].fillna(df['Age'].median())
```

```python



# Fix Education (0 is invalid)

df.loc[df['Education'] == 0, 'Education'] = np.nan
df['Education'] = df['Education'].fillna(df['Education'].median())


# Fix Rating / Scale Columns

rating_cols = [
    'EnvironmentSatisfaction',
    'JobSatisfaction',
    'RelationshipSatisfaction',
    'WorkLifeBalance',
    'JobInvolvement'
]

for col in rating_cols:
    df.loc[df[col] == 0, col] = np.nan
    df[col] = df[col].fillna(df[col].median())


#  EmployeeNumber Handling 
df.loc[df['EmployeeNumber'] == 0, 'EmployeeNumber'] = np.nan
df['EmployeeNumber'] = df['EmployeeNumber'].fillna(method='ffill')


print("Remaining Missing Values →", df.isna().sum().sum())
```
